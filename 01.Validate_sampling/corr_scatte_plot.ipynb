```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

```python
expLevel_table = pd.read_csv("../cuffdiff/exp_table.csv")
```

```python
expLevel_table = expLevel_table.drop(['tracking_id'],axis=1)
expLevel_table
```

```python
log_expLevel_table = expLevel_table.drop(['tracking_id'],axis=1)
log_expLevel_table = np.log2(log_expLevel_table + 1) # inf을 막기 위해서 +1
log_expLevel_table
```

```python
corr = expLevel_table.corr()
corr
```

```python
def plotLogExpLevel (gene1, gene2):
    #sns.set(rc = {'figure.figsize':(12,12)})
    gene1_csv = expLevel_table[gene1]
    gene2_csv = expLevel_table[gene2]
    plot = sns.scatterplot(x=gene1,y=gene2,data=log_expLevel_table)
    corr_table = pd.concat([gene1_csv,gene2_csv],axis=1)
    # text box
    box = {'boxstyle':'round','ec': (1.0, 0.5, 0.5),'fc': (1.0, 0.8, 0.8)}
    corr = corr_table.corr()
    corr = corr[gene1][1]
    my_txt = "Correlation = " + str(corr)
    plt.text(2,15,my_txt,bbox=box)
    plt.title(gene1 + " and " + gene2 + "gene exp level")
    save_file_name = gene1 + "_and_" + gene2 + "gene_exp_level"
    plt.savefig(save_file_name)
    #plt.text(1000,9000,my_txt,bbox=box)
```

## corr graph

```python
plotLogExpLevel("BI","RI") 
```

```python
plotLogExpLevel("BI","GW")
```

```python
plotLogExpLevel("RR","BR")
```
