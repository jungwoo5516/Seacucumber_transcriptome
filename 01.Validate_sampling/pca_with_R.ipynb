```python
import pandas as pd
```

```python
expLevel_PCA = pd.read_csv("./fpkm_PCA_valueTable.csv")
```

```python
expLevel_PCA = expLevel_PCA[['Dim.1','Dim.2']]
```

```python
samples = ['BI','BM','BR','BT','BW','GI','GM','GR','GT','GW','RI','RM','RR','RT','RW']
colors = ['Black','Black','Black','Black','Black','Green','Green','Green','Green','Green',
          'Red','Red','Red','Red','Red']
tissues = ['Intestion','Muscle','Respiratory','Tentacle','Wall','Intestion','Muscle',
           'Respiratory','Tentacle','Wall','Intestion','Muscle','Respiratory','Tentacle','Wall']

expLevel_PCA = pd.DataFrame(expLevel_PCA)
expLevel_PCA.index = samples
expLevel_PCA['samples'] =samples
expLevel_PCA['colors'] =colors
expLevel_PCA['tissues'] = tissues
expLevel_PCA
```

```python
expLevel_PCA.rename(columns={'Dim.1':'PC0'},inplace=True)
expLevel_PCA.rename(columns={'Dim.2':'PC1'},inplace=True)
expLevel_PCA
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plotPCAExpLevel (x, y, data, samplesOrColors):
    sns.set(rc = {'figure.figsize':(12,12)})
    
    if samplesOrColors == 0 :
        plot = sns.scatterplot(x=x,y=y,data=data,hue='samples',style='tissues',s=150)
        
    else :
        colors = ['#000000','g','r']
        sns.set_palette(sns.color_palette(colors))
        plot = sns.scatterplot(x=x,y=y,data=data,hue='colors',style='tissues',s=150)
        
    plt.title("PC1 and PC2 gene exp level")
    plt.savefig("PCA_graph")
    
```

```python
pca_info = pd.read_csv("./fpkm_PCA_explainTable.csv")
pca_info['percentage of variance'] # 분산도
```

```python
PCA_graph = plotPCAExpLevel('PC0','PC1',expLevel_PCA,1)
```
