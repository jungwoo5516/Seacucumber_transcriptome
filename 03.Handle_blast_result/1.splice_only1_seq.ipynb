```python
import pandas as pd
```

```python
blast_result = pd.read_csv('../00.blast_result/blastp_result.csv',names=['qseqid','sseqid','pident','length','mismatch','gapopen','qstart','qend','sstart',
                                  'send','evalue','bitscore'],delimiter='\t')
```

```python
blast_result
```

```python
new_blast_index = 0
new_blast_result = pd.DataFrame({'qseqid':[],'sseqid':[],'pident':[],'length':[],'mismatch':[],
                               'gapopen':[],'qstart':[],'qend':[],'sstart':[],'send':[],'evalue':[],'bitsocre':[]})
```

```python
new_blast_result.loc[0] = blast_result.loc[0]
```

```python
new_blast_result
```

```python
for i in range(len(blast_result)) :
    swiss_prot_ID = blast_result['sseqid'][i] # blast result를 한 line씩 읽는다.
    
    if swiss_prot_ID != new_blast_result['sseqid'][0] : # 만약 라인이 들어가 있지 않으면
        new_blast_result.loc[new_blast_index] = blast_result.loc[i] # 삽입을 한다.
    
    if i == len(blast_result)-1 : break

    if blast_result['sseqid'][i+1] != swiss_prot_ID:
        new_blast_index += 1
```

```python
new_blast_result
```

```python
new_blast_result.to_csv('blast_1vs1_matching.csv')
```
