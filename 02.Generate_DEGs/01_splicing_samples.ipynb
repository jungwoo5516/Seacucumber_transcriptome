```python
import pandas as pd
```

```python
diff_file = pd.read_csv('./gene_exp_spliced_diff2.csv')
```

```python
diff_file = diff_file.drop(['Unnamed: 0'],axis=1)
```

```python
diff_file
```

```python
def splicingSamples(sample1, sample2):
    # diff file sample1,2 목록을 추출
    sample1_list = (diff_file['sample_1'] == sample1)
    sample2_list = (diff_file['sample_2'] == sample2)
    
    # sample1과 2가 같이 있는 index를 저장하기 위한 list
    common_indexList = []
    
    for index in range(len(sample1_list)) :
        if (sample1_list[index] == True) and (sample2_list[index]==True):
            common_indexList.append(index)
    
    # diff file에서 위에서 추출한 index들에 해당하는 정보를 가져오기
    common_sampleList = []
    common_sampleList = diff_file.loc[common_indexList[0] : common_indexList[len(common_indexList)-1]]
    return common_sampleList
        
```

```python
def buildCompareFile(sample1,sample2):
    compare_file = splicingSamples(sample1,sample2)
    compare_file.to_csv((sample1 + "_and_" + sample2 + "_expressionLevel.csv"),index=False)
```

```python
buildCompareFile('BI','RI')
buildCompareFile('BI','GI')
buildCompareFile('GI','RI')
```

```python
buildCompareFile('BM','RM')
buildCompareFile('BM','GM')
buildCompareFile('GM','RM')
```

```python
buildCompareFile('BT','RT')
buildCompareFile('BT','GT')
buildCompareFile('GT','RT')
```
