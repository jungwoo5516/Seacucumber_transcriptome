```python
import pandas as pd
```

```python
swiss_prot_list = pd.read_csv('../../07.Blast/01.Handle_blast_result/blast_1vs1_matching.csv')
```

```python
swiss_prot_list
```

```python
# swiss_prot protein listd의 qseqid를 sample들과 비교하기 위해서 이름을 evm.TU~로 통일 시켜준다.
for i in range(len(swiss_prot_list)):
    tmp = swiss_prot_list['qseqid'][i]
    tmp = tmp.replace('model','TU')
    swiss_prot_list['qseqid'][i] = tmp
```

```python
swiss_prot_list = swiss_prot_list.drop(['Unnamed: 0'],axis=1)
```

```python
swiss_prot_list
```

```python
annotation_list = list(swiss_prot_list['qseqid'])
```

```python
len(annotation_list)
```

```python
DEG_file = pd.read_csv('./DEGs_list.csv')
DEG_list = list(DEG_file['gene'])
len(DEG_list)
```

```python
annotation_DEGs_list = []

for i in range(len(annotation_list)):
    if annotation_list[i] in DEG_list :
        annotation_DEGs_list.append(annotation_list[i]) 
```

```python
annotation_DEGs_list
```

```python
annotation_DEGs_list = pd.DataFrame(annotation_DEGs_list)
```

```python
annotation_DEGs_list.to_csv('annotation_DEGs_list.csv',index=False,header=False)
```
