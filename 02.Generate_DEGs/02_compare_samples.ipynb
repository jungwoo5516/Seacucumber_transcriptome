```python
def compareSamplesValue(data,minExpressionLevel,foldChangeValue) :
    temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],
                         'value_1':[],'value_2':[],'value_2':[],'significant':[]})
    
    for index in range(len(data)):
        sampel1_expressionLevel = data['value_1'][index]
        sampel2_expressionLevel = data['value_2'][index]
    
        # 최소 발현량 cutoff
        if (sampel1_expressionLevel > minExpressionLevel) and (sampel2_expressionLevel > minExpressionLevel):
            # fold change cutoff
            if (sampel1_expressionLevel/sampel2_expressionLevel>foldChangeValue) or (sampel2_expressionLevel/sampel1_expressionLevel>foldChangeValue):
                temp_list.loc[len(temp_list)] = data.loc[index]
    
    return temp_list
            
# 개수가 수 백개 정도 되는 cutoff를 찾기
```

```python
def buildSpecificGenesFile(data,minExpressionLevel,foldChangeValue) :
    new_file = compareSamplesValue(data,minExpressionLevel,foldChangeValue)
    sample1 = data['sample_1'][0]
    sample2 = data['sample_2'][0]
    new_file.to_csv((sample1 + "_and_" + sample2 + "_siginificantGenes.csv"),index=False)
```
