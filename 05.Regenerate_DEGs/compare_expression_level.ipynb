```python
# foldChange를 비교하는 functioin
def compareSamplesValue(data,minExpressionLevel,foldChangeValue) :
    temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],
                         'value_1':[],'value_2':[],'significant':[]})
    
    for index in range(len(data)):
        sampel1_expressionLevel = data['value_1'][index]
        sampel2_expressionLevel = data['value_2'][index]
    
        # 최소 발현량 cutoff
        if (sampel1_expressionLevel > minExpressionLevel) and (sampel2_expressionLevel > minExpressionLevel):
            # fold change cutoff
            # sample1이 더 크게 발현하는 것을 찾는다.
            if (sampel1_expressionLevel/sampel2_expressionLevel>foldChangeValue) :
                temp_list.loc[len(temp_list)] = data.loc[index]
    
    return temp_list
```

```python
# foldchange가 10이 넘는것들 중에서 공통되는 gene list
def findCommonGenes(data1,data2):
    common_genes = set(data1['gene']) & set(data2['gene'])
    
    # temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],'value_1':[],'value_2':[],'value_2':[],'significant':[]})
    
    return common_genes
```

```python
BI_siginificant_genes = findCommonGenes(BI_RI_foldChange_10,BI_GI_foldChange_10)
```

```python
# BI sinificant한 list에 있지 않는 gene들을 없앤다.

def buildCommonGeneDataFrame (gene_list, data):
    
    index = 0
    temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],'value_1':[],'value_2':[],'value_2':[],'significant':[]})
        
    for i in range(len(data)):
        if data['gene'][i] in gene_list :
            #data = data.drop(index=i)
            temp_list.loc[index] = data.loc[i]
            index += 1
    
    return temp_list
```

```python
BI_GI_siginificant_genes = buildCommonGeneDataFrame(BI_siginificant_genes,BI_GI_foldChange_10)
BI_RI_siginificant_genes = buildCommonGeneDataFrame(BI_siginificant_genes,BI_RI_foldChange_10)
```

```python
# BI의 DEG목록을 뽑기 위해서 위의 공통되지 않는 gene들을 제거한 DataFrame을 만든다.
def concatCommonGeneDataFrame (data1,data2):
    
    index = 0
    temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],'sample_3':[],'value_1':[],'value_2':[],'value_3':[]})
    
    for i in range(len(data1)):
        temp_line = [data1['gene'][i],      # gene
                     data1['sample_1'][i],  # sample_1
                     data1['sample_2'][i],  # sample_2
                     data2['sample_2'][i],  # sample_3 * data2사용
                     data1['value_1'][i],   # value_1
                     data1['value_2'][i],   # value_2
                     data2['value_2'][i],   # value_3 * data2사용
                     ]
        temp_list.loc[i] = temp_line
    
    return temp_list
        
```

```python
BI_DEGs = concatCommonGeneDataFrame(BI_GI_siginificant_genes,BI_RI_siginificant_genes)
```

```python
def buildDEGsPipeline (data1,data2,minExpressionLevel,foldChangeValue) :
    data1_foldChange_10 = compareSamplesValue(data1,minExpressionLevel,foldChangeValue)
    data2_foldChange_10 = compareSamplesValue(data2,minExpressionLevel,foldChangeValue)
 
    data_common_genes = findCommonGenes(data1_foldChange_10,data2_foldChange_10)    
    data1_significant_genes = buildCommonGeneDataFrame(data_common_genes,data1_foldChange_10)
    data2_significant_genes = buildCommonGeneDataFrame(data_common_genes,data2_foldChange_10)
    
    dataDEGs = concatCommonGeneDataFrame(data1_significant_genes,data2_significant_genes)
    
    return dataDEGs    
```

```python
# foldchange가 10이 넘는것들 중에서 공통되는 gene list
def findCommonGenes(data1,data2):
    common_genes = set(data1['gene']) & set(data2['gene'])
    
    # temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],'value_1':[],'value_2':[],'value_2':[],'significant':[]})
    
    return common_genes
```

```python
# BI sinificant한 list에 있지 않는 gene들을 없앤다.

def buildCommonGeneDataFrame (gene_list, data):
    
    index = 0
    temp_list = pd.DataFrame({'gene':[],'sample_1':[],'sample_2':[],'value_1':[],'value_2':[],'value_2':[],'significant':[]})
        
    for i in range(len(data)):
        if data['gene'][i] in gene_list :
            #data = data.drop(index=i)
            temp_list.loc[index] = data.loc[i]
            index += 1
    
    return temp_list
```

```python
def buildDEGsPipeline (data1,data2,minExpressionLevel,foldChangeValue, color) :
    
    if color == "black":
        data1_foldChange_10 = compareSamplesValue(data1,minExpressionLevel,foldChangeValue)
        data2_foldChange_10 = compareSamplesValue(data2,minExpressionLevel,foldChangeValue)
        
    elif color == "green" :
        data1_foldChange_10 = compareSamplesValue2(data1,minExpressionLevel,foldChangeValue)
        data2_foldChange_10 = compareSamplesValue(data2,minExpressionLevel,foldChangeValue)
    
    elif color == "red" :
        data1_foldChange_10 = compareSamplesValue2(data1,minExpressionLevel,foldChangeValue)
        data2_foldChange_10 = compareSamplesValue2(data2,minExpressionLevel,foldChangeValue)
    
    else :
        print("input color black or red or green")
     
    data_common_genes = findCommonGenes(data1_foldChange_10,data2_foldChange_10)    
    data1_significant_genes = buildCommonGeneDataFrame(data_common_genes,data1_foldChange_10)
    data2_significant_genes = buildCommonGeneDataFrame(data_common_genes,data2_foldChange_10)
    
    if color == 'red':
        dataDEGs = concatCommonGeneDataFrame2(data1_significant_genes,data2_significant_genes)
    
    else :
        dataDEGs = concatCommonGeneDataFrame(data1_significant_genes,data2_significant_genes)

    return dataDEGs    
```
