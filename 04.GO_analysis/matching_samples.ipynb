## Function: matching samples and swiss prot protein ID


```python
def matchSamplesAndSwissprotID (samples_name, blast_result, tissue_or_color) : # samples_name은 str로 blast_result는 DataFrame의 형식으로 input해준다.
    samples_path_name = samples_name.replace('_','_and_') # ex) BI_GI가 input인데 이 이름을 BI_and_GI로 변경
    
    if tissue_or_color == None :
        samples_path = '../../06.generate_DEGs/splice_by_sameTissue/' + samples_path_name + '_siginificantGenes.csv'
    elif tissue_or_color == "black":
        samples_path = '../../06.generate_DEGs/splice_by_sameColor/black/' + samples_path_name + '_siginificantGenes.csv'
    elif tissue_or_color == "red":
        samples_path = '../../06.generate_DEGs/splice_by_sameColor/red/' + samples_path_name + '_siginificantGenes.csv'
    elif tissue_or_color == "green":
        samples_path = '../../06.generate_DEGs/splice_by_sameColor/green/' + samples_path_name + '_siginificantGenes.csv'
        
    sample = pd.read_csv(samples_path) # sample을 load
    
    # blast 결과와 matching 되는 Protein list
    matcing_protein_df = pd.DataFrame({'qseqid':[],'sseqid':[],'pident':[],'length':[],'mismatch':[],
                               'gapopen':[],'qstart':[],'qend':[],'sstart':[],'send':[],'evalue':[],'bitsocre':[]})
    
    # for문에서 비교를 하기위해서 list로 변형
    samples_gene_list = list(sample['gene']) 
    
    # 위의 matching된 protein list의 index
    index = 0
    for i in range(len(swiss_prot_list)) :
        if blast_result['qseqid'][i] in samples_gene_list :
            matcing_protein_df.loc[index] = blast_result.loc[i]
            index += 1

    output_name = samples_name + '_proteinList'
    matcing_protein_df['sseqid'].to_csv(output_name,index=False)
    return matcing_protein_df['sseqid'] 
```
